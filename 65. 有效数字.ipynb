## 56. 合并区间

给出一个区间的集合，请合并所有重叠的区间。

示例 1:

输入: [[1,3],[2,6],[8,10],[15,18]]
输出: [[1,6],[8,10],[15,18]]
解释: 区间 [1,3] 和 [2,6] 重叠, 将它们合并为 [1,6].
示例 2:

输入: [[1,4],[4,5]]
输出: [[1,5]]
解释: 区间 [1,4] 和 [4,5] 可被视为重叠区间。



```c++
class Solution {
public:
    vector<vector<int>> merge(vector<vector<int>>& intervals) {
        if(intervals.empty())
            return {};
        sort(intervals.begin(),intervals.end());   //先排序
        vector<vector<int>> res{intervals[0]};
        for(int i=1;i<intervals.size();++i){
            if(res.back()[1]<intervals[i][0]){      //当区间的左边界大于上一个区间的右边界时，将该区间加入结果
                res.push_back(intervals[i]);        
            }
            else{
                res.back()[1]=max(res.back()[1],intervals[i][1]);     //小于的，比较两个区间的右边界，选择较大的边界
            }
            
        }
        return res;
        
    }
};
```

## 60. 第k个排列

给出集合 [1,2,3,…,n]，其所有元素共有 n! 种排列。

按大小顺序列出所有排列情况，并一一标记，当 n = 3 时, 所有排列如下：

"123"
"132"
"213"
"231"
"312"
"321"
给定 n 和 k，返回第 k 个排列。

说明：

给定 n 的范围是 [1, 9]。
给定 k 的范围是[1,  n!]。
示例 1:

输入: n = 3, k = 3
输出: "213"
示例 2:

输入: n = 4, k = 9
输出: "2314"



```c++
class Solution {
public:
    string getPermutation(int n, int k) {//需要找到排列的规律
        string res;
        string num="123456789";
        vector<int> f(n,1);
        for(int i=1;i<n;++i)
            f[i]=f[i-1]*i;
            --k;
        for(int i=n;i>=1;--i){
            int j=k/f[i-1];       
            k%=f[i-1];
            res.push_back(num[j]);
            num.erase(j,1);
        }
        return res;
    }
};
```

## 65. 有效数字

验证给定的字符串是否可以解释为十进制数字。

例如:

"0" => true
" 0.1 " => true
"abc" => false
"1 a" => false
"2e10" => true
" -90e3   " => true
" 1e" => false
"e3" => false
" 6e-1" => true
" 99e2.5 " => false
"53.5e93" => true
" --6 " => false
"-+3" => false
"95a54e53" => false

说明: 我们有意将问题陈述地比较模糊。在实现代码之前，你应当事先思考所有可能的情况。这里给出一份可能存在于有效十进制数字中的字符列表：

数字 0-9
指数 - "e"
正/负号 - "+"/"-"
小数点 - "."
当然，在输入中，这些字符的上下文也很重要。



```c++
class Solution {
public:
    bool isNumber(string s) {
         int len = s.size();
        int left = 0, right = len - 1;
        bool eExisted = false,dotExisted = false, digitExisited = false;
        while (s[left] == ' ') ++left;
        while (s[right] == ' ') --right;//跳过左右空格
        if (left >= right && (s[left] < '0' || s[left] > '9')) return false;  //如果只有一个非数字字符或者没有字符返回false
        //对首字符作判断，若为dot 标记，为数字作标记，若还不是+或-则返回false
        if (s[left] == '.') dotExisted = true;
        else if (s[left] >= '0' && s[left] <= '9') digitExisited = true;     
        else if (s[left] != '+' && s[left] != '-') return false;    
        //对中间字段处理
        for (int i = left + 1; i <= right - 1; ++i) {
            if (s[i] >= '0' && s[i] <= '9') digitExisited = true;//若为数字，标记 
            else if (s[i] == 'e' || s[i] == 'E') { //若为E
                if (!eExisted && s[i - 1] != '+' && s[i - 1] != '-' && digitExisited) eExisted = true;//E之前不能为+，-必须是数字
                else return false;  //若第二个字符不是数字和E||e则返回false
            } 
            else if (s[i] == '+' || s[i] == '-') { //若为+-，则之前必须为E
                if (s[i - 1] != 'e' && s[i - 1] != 'E') return false;                  
            } 
            else if (s[i] == '.') { //若为dot，必须是第一次出现且没有E
                if (!dotExisted && !eExisted) dotExisted = true;
                else return false;
            }
            else return false;
        }
        //对尾部字段处理，最后一个字符只能为数字或dot
        if (s[right] >= '0' && s[right] <= '9') return true;   
        else if (s[right] == '.' && !dotExisted && !eExisted && digitExisited) return true; //若为dot,前面必须没有dot,且必须是数字没有E
        else return false;
    }
};
```
